```python
from pydub import AudioSegment

def convert_to_m4a(input_file: str, output_file: str) -> None:
    """
    Convert an audio file to M4A format.

    Args:
        input_file (str): Path to the source audio file (e.g., .wav, .mp3).
        output_file (str): Path to the target M4A file.
    """
    audio = AudioSegment.from_file(input_file)
    audio.export(output_file, format="ipod")  # 'ipod' is the format name for .m4a

# Example usage
convert_to_m4a("yoruba_short_1.opus", "yoruba_short_1.wav")
```
